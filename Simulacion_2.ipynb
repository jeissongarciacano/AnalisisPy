# Simulación de precios de activos

```python
ticker = 'GE'
data = pd.DataFrame()
data[ticker] = yf.download(ticker, start = '2019-01-01')['Adj Close']
```

```python
#Rendimiento logarítmico
#pct_change() calcula el retorno simple de un conjunto de datos
rendimientoLogaritmico = np.log(1 + data.pct_change())
rendimientoLogaritmico.head()
```

```python
rendimientoLogaritmico.tail()
```

```python
u = rendimientoLogaritmico.mean()
u
```

```python
var = rendimientoLogaritmico.var()
var
```

```python
tendencia = u - (0.5 * var)
tendencia
```

```python
desv = rendimientoLogaritmico.std()
desv
```

```python
#Estadístico de prueba al 95%
norm.ppf(0.95)
```

```python
#Simulación
numeroIntervalos = 1000
iteraciones = 10
```

```python
retornosDiarios = np.exp(tendencia.values + desv.values * norm.ppf(np.random.rand(numeroIntervalos, iteraciones)))
retornosDiarios
```

```python
P0 = data.iloc[-1]
P0
```

```python
listaPrecios = np.zeros_like(retornosDiarios)
listaPrecios
```

```python
listaPrecios[0] = P0
listaPrecios
```

```python
for t in range(1, numeroIntervalos):
    listaPrecios[t] = listaPrecios[t - 1] * retornosDiarios[t]
listaPrecios
```

```python
plt.figure(figsize =(16,8))
plt.plot(listaPrecios)
plt.show()
```

# Cálculo de precios de instrumentos derivados (opciones)

```python
#Función para d1
def d1 (S, K, r, stdev, T):
    return (np.log(S/K) + (r + stdev ** 2 / 2) * T)/(stdev * np.sqrt(T))
```

```python
#Función para d2
def d2 (S, K, r, stdev, T):
    return (np.log(S/K) + (r - stdev ** 2 / 2) * T)/(stdev * np.sqrt(T))    
```

```python
norm.cdf(0)
```

```python
#Función para la fórmula de Black - Scholes
def BSF (S, K, r, stdev, T):
    return (S * norm.cdf(d1(S, K, r, stdev, T))) - (K * np.exp(-r * T) * norm.cdf(d2(S, K, r, stdev, T)))
```

```python
ticker = 'PG'
data = pd.DataFrame()
data[ticker] = yf.download(ticker, start = '2019-01-01')['Adj Close']
```

```python
S = data.iloc[-1]
S
```

```python
retornoLogaritmico = np.log(1 + data.pct_change())
stdev = retornoLogaritmico.std() * 250 **0.5
stdev
```

```python
r = 0.025
K = 110.0
T = 1
```

```python
#El valor que se va a recibir en caso que se ejecute la opción
d1(S, K, r, stdev, T)
```

```python
#El valor que se paga si se ejecuta la opción
d2(S, K, r, stdev, T)
```

```python
#Diferencia entre el valor a recibir y el valor a pagar si se ejecuta la opción
BSF(S, K, r, stdev, T)
```

# Simulación Monte Carlo para cálculo de derivados

```python
retornoLogaritmico = np.log(1 + data.pct_change())
```

```python
#Tasa Libre de riesgo
r = 0.025
```

```python
#Desviación estándar anual
desv = retornoLogaritmico.std() * 250 ** 0.5
desv
```

```python
type(desv)
```

```python
desv = desv.values
desv
```

```python
#Periodo de evaluación
T = 1.0
#Intervalos fijos
tIntervalos = 250
#Cambio fijo
deltaT = T / tIntervalos
iteraciones = 10000
```

```python
Z = np.random.standard_normal((tIntervalos + 1, iteraciones))
S = np.zeros_like(Z)
S0 = data.iloc[-1]
S[0] = S0
S
```

```python
for t in range(1, tIntervalos + 1):
    S[t] = S[t - 1] * np.exp(((r - 0.5 * desv ** 2) * deltaT) + (desv * np.sqrt(deltaT) * Z[t]))
```

```python
S
```

```python
S.shape
```

```python
plt.figure(figsize = (16,8))
plt.plot(S)
plt.show()
```

```python
plt.figure(figsize = (16,8))
plt.plot(S[:,:10])
plt.show()
```

```python
#Determinación de la opción
# Si S - K > 0 opción de compra
# Si S - K < 0 no opción de compra
p = np.maximum(S[-1] - K,0)
#p = S[-1] - K
#maximum crea una matriz de 0 o de la diferencia, dependiendo de cuál es máxima
p
```

```python
p.shape
```

```python
C = np.exp(-r * T) * np.sum(p) / iteraciones
C
```
